```python
# ğŸŒ± Smart Waste Classification using CNN
# Author: Maheshwari Mahendra Gaikwad
# Domain: Sustainability
# College: AISSMS College of Engineering

# Step 1: Install Kaggle
!pip install kaggle

# Step 2: Upload Kaggle API key
from google.colab import files
files.upload()  # Upload your kaggle.json

# Step 3: Move API key to correct folder
!mkdir -p ~/.kaggle
!cp kaggle.json ~/.kaggle/
!chmod 600 ~/.kaggle/kaggle.json

# Step 4: Download the Garbage Classification Dataset from Kaggle (force overwrite)
!kaggle datasets download -d mostafaabla/garbage-classification --force

# Step 5: Extract dataset automatically and overwrite existing files
!unzip -o -q garbage-classification.zip -d /content/dataset

# Step 6: Verify folder structure
import os
dataset_path = '/content/dataset/garbage_classification'
print("Classes in dataset:", os.listdir(dataset_path))

# Step 7: Data preprocessing with automatic train/validation split
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_datagen = ImageDataGenerator(
    rescale=1./255,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    validation_split=0.2  # 20% for validation
)

train_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical',
    subset='training',  # 80% training
    shuffle=True
)

validation_generator = train_datagen.flow_from_directory(
    dataset_path,
    target_size=(128,128),
    batch_size=32,
    class_mode='categorical',
    subset='validation',  # 20% validation
    shuffle=False
)

# Step 8: Build CNN model for 12 classes
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense, Dropout

model = Sequential([
    Conv2D(32, (3,3), activation='relu', input_shape=(128,128,3)),
    MaxPooling2D(2,2),
    Conv2D(64, (3,3), activation='relu'),
    MaxPooling2D(2,2),
    Flatten(),
    Dense(128, activation='relu'),
    Dropout(0.5),
    Dense(12, activation='softmax')  # 12 classes
])

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
model.summary()

# Step 9: Train the model
history = model.fit(
    train_generator,
    epochs=10,  # Increase for better accuracy
    validation_data=validation_generator
)

# Step 10: Evaluate the model
loss, accuracy = model.evaluate(validation_generator)
print(f"âœ… Validation Accuracy: {accuracy*100:.2f}%")

# Step 11: Plot training and validation accuracy
import matplotlib.pyplot as plt
plt.plot(history.history['accuracy'], label='Train Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Model Accuracy')
plt.xlabel('Epochs')
plt.ylabel('Accuracy')
plt.legend()
plt.show()

# Step 12: Confusion matrix and classification report
from sklearn.metrics import classification_report, confusion_matrix
import numpy as np
import seaborn as sns

Y_pred = model.predict(validation_generator)
y_pred = np.argmax(Y_pred, axis=1)

print('Confusion Matrix')
cm = confusion_matrix(validation_generator.classes, y_pred)
sns.heatmap(cm, annot=True, fmt='d', cmap='Greens')
plt.show()

print('Classification Report')
print(classification_report(validation_generator.classes, y_pred, target_names=list(train_generator.class_indices.keys())))

# Step 13: Save the trained model
model.save('waste_classification_model.h5')
print("âœ… Model saved as waste_classification_model.h5")

# Step 14: Optional - Test with new image
from tensorflow.keras.preprocessing import image

# img = image.load_img('/content/test_image.jpg', target_size=(128,128))
# img_array = image.img_to_array(img)/255.0
# img_array = np.expand_dims(img_array, axis=0)
# prediction = np.argmax(model.predict(img_array))
# classes = list(train_generator.class_indices.keys())
# print("Predicted Waste Type:", classes[prediction])
```
